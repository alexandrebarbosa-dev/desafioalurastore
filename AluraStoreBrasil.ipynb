### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head(25)
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja1['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
```

```python
resumo_faturamento = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento': [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
})
resumo_faturamento
```

# 2. Vendas por Categoria

```python
Vendas_produtos_categoria1 = loja1['Categoria do Produto'].value_counts()
Vendas_produtos_categoria2 = loja2['Categoria do Produto'].value_counts()
Vendas_produtos_categoria3 = loja3['Categoria do Produto'].value_counts()
Vendas_produtos_categoria4 = loja4['Categoria do Produto'].value_counts()
```

# 3. Média de Avaliação das Lojas


```python
media_avaliacao1 = loja1['Avaliação da compra'].mean()
media_avaliacao2 = loja2['Avaliação da compra'].mean()
media_avaliacao3 = loja3['Avaliação da compra'].mean()
media_avaliacao4 = loja4['Avaliação da compra'].mean()
```

```python
resumo_avaliacao = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Média de Avaliação': [media_avaliacao1, media_avaliacao2, media_avaliacao3, media_avaliacao4]
})
resumo_avaliacao

```

```python
lojas = [loja1, loja2, loja3, loja4]

resumo = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento total': [loja['Preço'].sum() for loja in lojas],
    'Média de Avaliação': [loja['Avaliação da compra'].mean() for loja in lojas]
})
resumo
```

```python
from matplotlib import axes
def grafico_vendas_categoria():

  lojas = [loja1, loja2, loja3, loja4]
  nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

  fig, axes = plt.subplots(1, 4, figsize=(20,5))

  for i in range(4):
      lojas[i]['Categoria do Produto'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[i])
      axes[i].set_title(nomes[i])
      axes[i].set_ylabel('')

  plt.tight_layout()
  plt.show()
```

```python
grafico_vendas_categoria()
```

```python
def grafico_media_avaliacao():

    plt.figure(figsize=(8, 4))

    barras_avaliacao = plt.bar(
        resumo_avaliacao['Loja'],
        resumo_avaliacao['Média de Avaliação']
    )

    plt.title('Média de Avaliação por Loja')
    plt.xlabel('Loja')
    plt.ylabel('Média de Avaliação')

    for barra in barras_avaliacao:
        altura = barra.get_height()
        plt.text(
            barra.get_x() + barra.get_width() / 2,
            altura,
            f'{altura:.2f}',
            ha='center',
            va='bottom'
        )


    plt.show()
```

```python
grafico_media_avaliacao()
```

```python
# Gráfico de avaliação por linha como opção

def grafico_media_avaliacao_linha():

    plt.figure(figsize=(10, 6))

    x = resumo_avaliacao['Loja']
    y = resumo_avaliacao['Média de Avaliação']

    plt.plot(x, y, marker='o')

    for i in range(len(x)):
        plt.text(
            x[i],
            y[i] + 0.02,
            f'{y[i]:.2f}',
            ha='center',
            va='bottom'
        )

    plt.title('Média de Avaliação por Loja')
    plt.xlabel('Loja')
    plt.ylabel('Média de Avaliação')
    plt.ylim(0, 5.2)
    plt.grid(True)

    plt.show()
```

```python
grafico_media_avaliacao_linha()
```

```python
import matplotlib.pyplot as plt

def grafico_column_line():

    fig, ax1 = plt.subplots(figsize=(10,6))

    # Barras -> Faturamento
    ax1.bar(
        resumo_faturamento['Loja'],
        resumo_faturamento['Faturamento'],
        alpha=0.7
    )
    ax1.set_xlabel('Loja')
    ax1.set_ylabel('Faturamento')

    # Linha -> Média de Avaliação
    ax2 = ax1.twinx()

    ax2.plot(
        resumo_avaliacao['Loja'],
        resumo_avaliacao['Média de Avaliação'],
        marker='o'
    )
    ax2.set_ylabel('Média de Avaliação')

    plt.title('Faturamento x Média de Avaliação por Loja')
    plt.show()
```

```python
grafico_column_line()
```

```python
Lojas = ['loja1', 'loja2', 'loja3', 'loja4']
Frete = [media_frete1, media_frete2, media_frete3, media_frete4]

def grafico_custo_frete():

  plt.figure(figsize=(8,5))
  barras = plt.bar(Lojas, Frete)
  plt.title('Custo x Frete por Loja')
  plt.xlabel('Loja')
  plt.ylabel('Custo x Frete')

  for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        altura,
        f'{altura:.2f}',
        ha='center',
        va='bottom'
    )
  plt.show()



```

# 📊 Relatório de Análise – Alura Store (produzido com auxílio de IA a partir de observações minhas em estado bruto.

## 1. Introdução

Este relatório tem como objetivo analisar o desempenho das quatro unidades da rede Alura Store, a fim de auxiliar o Senhor João na decisão sobre qual loja deve ser vendida para viabilizar um novo empreendimento.

Para embasar a recomendação final, foram analisadas as seguintes métricas:

* Faturamento total por loja
* Média de avaliação dos clientes
* Distribuição das vendas por categoria
* Produtos mais e menos vendidos
* Custo médio de frete

A análise combinou dados quantitativos e visualizações gráficas para permitir uma comparação clara entre as unidades.

---

## 2. Análise dos Dados

### 2.1 Faturamento

A Loja 1 apresentou o maior faturamento entre as quatro unidades, demonstrando forte desempenho em volume financeiro.

Por outro lado, a Loja 4 registrou o menor faturamento do grupo, ficando abaixo das demais no comparativo geral.

Esse dado, isoladamente, poderia indicar a Loja 4 como candidata à venda. No entanto, é necessário analisar os demais indicadores antes de uma conclusão definitiva.

---

```python
grafico_faturamento()
```

### 2.2 Avaliação dos Clientes

Ao observar a média de avaliações:

* A Loja 1 apresentou a menor média de avaliação entre as quatro.
* A Loja 4 obteve avaliação superior à da Loja 1.

Esse resultado revela um ponto importante: embora a Loja 1 fature mais, ela apresenta menor satisfação dos clientes. Já a Loja 4, mesmo com faturamento inferior, mantém percepção mais positiva junto ao público.

Essa relação sugere que, neste caso especificamente, faturamento elevado não necessariamente está associado à melhor experiência do cliente.

---

```python
grafico_media_avaliacao()
```

### 2.3 Vendas por Categoria

A análise do gráfico de vendas por categoria demonstra grande semelhança entre as quatro lojas.

Não há concentração significativa ou distorções relevantes na distribuição das categorias vendidas. Todas apresentam comportamento bastante semelhante, indicando que:

* Não há vantagem competitiva clara em mix de produtos.
* As lojas operam com estrutura de portfólio equilibrada.

Portanto, esse critério não influencia de forma decisiva a escolha da loja a ser vendida.

---

```python
grafico_vendas_categoria()
```

### 2.4 Custo Médio de Frete

A Loja 1 apresentou o maior custo médio de frete, enquanto a Loja 4 apresentou o menor.

Entretanto, é importante destacar que o frete é custeado pelo cliente. A análise não demonstrou impacto direto desse fator no desempenho de vendas ou nas avaliações.

Dessa forma, embora seja um indicador relevante para compreensão operacional, o frete não se mostrou determinante para a decisão estratégica.

---

```python
grafico_custo_frete()
```

## 3. Conclusão e Recomendação

Ao combinar os indicadores analisados — especialmente o gráfico comparativo entre faturamento e média de avaliação — observa-se que:

* A Loja 1 possui o maior faturamento, porém apresenta a pior avaliação.
* A Loja 4 possui o menor faturamento, mas mantém nível de avaliação superior à Loja 1.
* As categorias vendidas apresentam comportamento semelhante entre todas as lojas.
* O custo de frete não demonstrou impacto relevante no desempenho.

Diante desse cenário, a recomendação é que a Loja 4 seja vendida.

Embora apresente avaliação satisfatória, seu desempenho financeiro é o menor entre as unidades analisadas. Considerando que as demais lojas apresentam faturamento superior e estrutura de vendas semelhante, a alienação da Loja 4 tende a gerar menor impacto financeiro para a rede como um todo.

Essa decisão preserva as unidades com maior geração de receita, mantendo a sustentabilidade e aprimoramento do negócio enquanto possibilita a realização do novo investimento pretendido pelo Senhor João.
